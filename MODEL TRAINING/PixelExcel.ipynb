```python
import numpy as np
import pandas as pd
import os
import cv2
```

```python
val = {}
```

```python
for plant in ['Strawberry','Peach','Potato','Corn','Pepper','Grape','Apple','Cherry','Tomato']:
  for i in os.listdir('/content/drive/MyDrive/'+plant+'/Valid'):
    df = len(os.listdir('/content/drive/MyDrive/'+plant+'/Valid/'+i))
    val [i] = df
  for i in os.listdir('/content/drive/MyDrive/'+plant+'/Train'):
    df = len(os.listdir('/content/drive/MyDrive/'+plant+'/Train/'+i))
    val [i] = val[i]+ df
```

```python
val
```

```python
name = list(val.keys())
test = []
train = []
value = list(val.values())
for i in value:
  test.append(int(i*0.33))
  train.append(int(i*0.67))
final = {'Name':name,
         'Training Images': train,
         'Testing Images': test}
final
```

```python
df = pd.DataFrame(final)
```
