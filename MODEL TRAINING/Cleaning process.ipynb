```python
import PyPDF2
import fitz  # PyMuPDF
import os
document1=''
document2 = ''
def extract_text_pdf(file_path):
    text = ""
    with open(file_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        for page_number in range(len(reader.pages)):
            text += reader.pages[page_number].extract_text()
    return text

def extract_text_pdf_mupdf(file_path):
    text = ""
    doc = fitz.open(file_path)
    for page_number in range(len(doc)):
        page = doc.load_page(page_number)
        text += page.get_text()
    return text

p = "/content/drive/MyDrive/Marathi Docs/"
for file in os.listdir("/content/drive/MyDrive/Marathi Docs"):
  if file.endswith(".PDF"):
    pdf_path=p+file
    extracted_text = extract_text_pdf(pdf_path)
    print("Extracted text using PyPDF2:")
    #print(extracted_text)
    document1 = document1 + extracted_text
    extracted_text_mupdf = extract_text_pdf_mupdf(pdf_path)
    print("\nExtracted text using PyMuPDF:")
    document2 = document2 + extracted_text_mupdf


```
